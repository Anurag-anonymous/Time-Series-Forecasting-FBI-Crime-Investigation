```python
import pandas as pd
from sklearn.metrics import mean_absolute_error, mean_squared_error
from statsmodels.tsa.seasonal import seasonal_decompose
import matplotlib.pyplot as plt
import numpy as np
from prophet import Prophet
path = "Train.xlsx"
df = pd.read_excel(path, nrows=None)
print("shape:",df.shape)
print("\ncolumns:",list(df.columns))

if 'Date' in df.columns:
    df['Date'] = pd.to_datetime(df['Date'], errors = 'coerce')
else:
    df['Date'] = pd.to_datetime(df[['YEAR','MONTH','DAY']])

daily_counts = df.groupby(df['Date'].dt.date).size()
monthly_counts = df.groupby([df['Date'].dt.to_period('M')]).size()
print("Daily series sample:\n", daily_counts.head())
print("\nMonthly series sample:\n", monthly_counts.head())


plt.figure(figsize=(12,4))
daily_counts.plot()
plt.title("Daily Crime Counts")
plt.xlabel("Date")
plt.ylabel("Number of Incidents")
plt.show()


plt.figure(figsize=(12,4))
monthly_counts.plot(kind='line')
plt.title("Monthly Crime Counts")
plt.xlabel("Month")
plt.ylabel("Number of Incidents")
plt.show()


# Convert monthly_counts to a proper time series with datetime index
monthly_series = monthly_counts.copy()
monthly_series.index = monthly_series.index.to_timestamp()

# Seasonal decomposition
result = seasonal_decompose(monthly_series, model='additive', period=12)

# Plot decomposition
result.plot()
plt.tight_layout()
plt.show()



# train-test split: keep last 12 months for testing (1 year)
train = monthly_series.iloc[:-12]
test = monthly_series.iloc[-0:]

print("Train period:", train.index.min(), "to", train.index.max())
print("Test period:", test.index.min(), "to", test.index.max())


# Naive forecast
naive_forecast = pd.Series(
    [train.iloc[-1]] * len(test),
    index=test.index
)

# Evaluation
mae = mean_absolute_error(test, naive_forecast)
rmse = np.sqrt(mean_squared_error(test, naive_forecast))
print(f"Naive Forecast -> MAE: {mae:.2f}, RMSE: {rmse:.2f}")

# Plot

plt.figure(figsize=(10,4))
plt.plot(train, label='Train')
plt.plot(test, label='Test', color='orange')
plt.plot(naive_forecast, label='Naive Forecast', color='green')
plt.title("Naive Forecast vs Actual")
plt.legend()
plt.show()


window = 3  # last 3 months average
moving_avg_forecast = pd.Series(
    [train.iloc[-window:].mean()] * len(test),
    index=test.index
)

mae_ma = mean_absolute_error(test, moving_avg_forecast)
rmse_ma = np.sqrt(mean_squared_error(test, moving_avg_forecast))
print(f"Moving Average (3-month) -> MAE: {mae_ma:.2f}, RMSE: {rmse_ma:.2f}")

# Plot comparison
plt.figure(figsize=(10,4))
plt.plot(train, label='Train')
plt.plot(test, label='Test', color='orange')
plt.plot(moving_avg_forecast, label='Moving Average Forecast', color='purple')
plt.legend()
plt.title("Moving Average Forecast vs Actual")
plt.show()


# 1️⃣ Prepare data
monthly_series = monthly_counts.copy()
monthly_series.index = monthly_series.index.to_timestamp()

prophet_df = monthly_series.reset_index()
prophet_df.columns = ['ds', 'y']  # Prophet requires ds (date) and y (target)

# 2️⃣ Split into train and test
train_df = prophet_df.iloc[:-12].copy()
test_df = prophet_df.iloc[-12:].copy()

# 3️⃣ Train Prophet model
m = Prophet(yearly_seasonality=True, weekly_seasonality=False, daily_seasonality=False)
m.fit(train_df)

# 4️⃣ Forecast next 12 months
future = m.make_future_dataframe(periods=12, freq='ME')  # 'M' = month end
forecast = m.predict(future)

# 5️⃣ Convert to month-end to match frequency
forecast['ds'] = pd.to_datetime(forecast['ds']) + pd.offsets.MonthEnd(0)
test_df['ds'] = pd.to_datetime(test_df['ds']) + pd.offsets.MonthEnd(0)

# 6️⃣ Set indices
forecast_df = forecast[['ds', 'yhat']].set_index('ds')
test_df = test_df.set_index('ds')

# 7️⃣ Align for common months
common_dates = forecast_df.index.intersection(test_df.index)
forecast_test = forecast_df.loc[common_dates, 'yhat']
actual_test = test_df.loc[common_dates, 'y']

# 8️⃣ Evaluate
if len(common_dates) > 0:
    mae_prophet = mean_absolute_error(actual_test, forecast_test)
    rmse_prophet = np.sqrt(mean_squared_error(actual_test, forecast_test))
    print(f"✅ Prophet Forecast -> MAE: {mae_prophet:.2f}, RMSE: {rmse_prophet:.2f}")
else:
    print("⚠️ No overlapping months found — check date alignment.")

# 9️⃣ Plot the forecast vs actual
plt.figure(figsize=(10,4))
plt.plot(train_df['ds'], train_df['y'], label='Train')
plt.plot(test_df.index, test_df['y'], label='Test', color='orange')
plt.plot(forecast_df.index, forecast_df['yhat'], label='Prophet Forecast', color='green')
plt.legend()
plt.title("Prophet Forecast vs Actual (Monthly Crime Counts)")
plt.show()

# 10️⃣ Optional: Prophet component plots
m.plot_components(forecast)


# Let's test different Prophet configurations
params = [
    {'changepoint_prior_scale': 0.05, 'seasonality_prior_scale': 10.0},  # default
    {'changepoint_prior_scale': 0.1, 'seasonality_prior_scale': 15.0},   # more flexible
    {'changepoint_prior_scale': 0.01, 'seasonality_prior_scale': 5.0},   # less flexible
]

results = []

for p in params:
    m = Prophet(
        yearly_seasonality=True,
        weekly_seasonality=False,
        daily_seasonality=False,
        changepoint_prior_scale=p['changepoint_prior_scale'],
        seasonality_prior_scale=p['seasonality_prior_scale']
    )
    m.fit(train_df)
    
    future = m.make_future_dataframe(periods=12, freq='ME')
    forecast = m.predict(future)
    
    forecast_df = forecast[['ds', 'yhat']].copy()
    forecast_df['ds'] = pd.to_datetime(forecast_df['ds']) + pd.offsets.MonthEnd(0)
    forecast_df.set_index('ds', inplace=True)
    
    common_dates = forecast_df.index.intersection(test_df.index)
    forecast_test = forecast_df.loc[common_dates, 'yhat']
    actual_test = test_df.loc[common_dates, 'y']
    
    mae = mean_absolute_error(actual_test, forecast_test)
    rmse = np.sqrt(mean_squared_error(actual_test, forecast_test))
    
    results.append({
        'changepoint_prior_scale': p['changepoint_prior_scale'],
        'seasonality_prior_scale': p['seasonality_prior_scale'],
        'MAE': mae,
        'RMSE': rmse
    })

results_df = pd.DataFrame(results)
print(results_df)



# Train with best parameters (replace with your best from above)
m_best = Prophet(
    yearly_seasonality=True,
    weekly_seasonality=False,
    daily_seasonality=False,
    changepoint_prior_scale= 0.05,  # update this
    seasonality_prior_scale=10.0  # update this
)
m_best.fit(prophet_df)

# Forecast for next 12 future months
future_full = m_best.make_future_dataframe(periods=12, freq='ME')
forecast_full = m_best.predict(future_full)

# Plot final forecast
m_best.plot(forecast_full)
plt.title("Future Crime Forecast (Next 12 Months)")
plt.show()

```
